```python
import numpy as np

# Create a 3x3 matrix
A = np.array([
    [2, 1, 3],
    [0, -1, 4],
    [5, 2, 0]
])

# Compute determinant
det_A = np.linalg.det(A)

print("Matrix A:\n", A)
print("Determinant of A:", det_A)
```

```python
import numpy as np

# Define the matrix
A = np.array([
    [2, 1, 3],
    [0, -1, 4],
    [5, 2, 0]
])

# Compute the inverse
A_inv = np.linalg.inv(A)

print("Matrix A:\n", A)
print("\nInverse of A:\n", A_inv)
```

```python
import numpy as np

# Coefficient matrix
A = np.array([
    [2, 3],
    [4, 5]
])

# Constant terms
B = np.array([10, 20])

# Solve the system
solution = np.linalg.solve(A, B)

x, y = solution

print("x =", x)
print("y =", y)
```
