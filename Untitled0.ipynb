```python
def bfs(graph, start):
    visited = [False] * len(graph)
    queue = []

    queue.append(start)
    visited[start] = True

    print("BFS Traversal:", end=" ")

    while queue:
        node = queue.pop(0)
        print(node, end=" ")

        for neighbour in graph[node]:
            if not visited[neighbour]:
                visited[neighbour] = True
                queue.append(neighbour)

# Define the graph (example adjacency list)
# For example, a graph with 6 nodes (0-5)
# 0 -> 1, 2
# 1 -> 3, 4
# 2 -> 5
# 3, 4, 5 have no outgoing edges
graph = [
    [1, 2],  # Node 0 is connected to 1 and 2
    [3, 4],  # Node 1 is connected to 3 and 4
    [5],     # Node 2 is connected to 5
    [],      # Node 3 has no connections
    [],      # Node 4 has no connections
    []       # Node 5 has no connections
]

# Run BFS
bfs(graph, 0)
```
