```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("cars.csv")

# Clean column names
data.columns = data.columns.str.strip().str.lower()

# Convert numeric columns properly
numeric_cols = ["mpg", "cylinders", "cubicinches", "hp", "weightlbs", "time-to-60", "year"]
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Remove rows with missing values
data.dropna(inplace=True)

print("First 5 rows of dataset:\n")
print(data.head())

# ----------------------------
# Filtering Data
# ----------------------------
# Cars with mileage greater than 25 mpg
filtered_data = data[data["mpg"] > 25]

print("\nFiltered Data (Cars with MPG > 25):\n")
print(filtered_data.head())

# ----------------------------
# Grouping & Aggregation
# ----------------------------
grouped_data = data.groupby("brand").agg({
    "mpg": ["mean", "min", "max"],
    "hp": ["mean", "min", "max"],
    "weightlbs": ["mean", "min", "max"],
    "brand": "count"
})

print("\nGrouped & Aggregated Statistics (By Brand):\n")
print(grouped_data)

# ----------------------------
# Overall Statistics
# ----------------------------
print("\nOverall Statistical Summary:\n")
print(data.describe())
```
