```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('/content/iris.csv')

# Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())
```

```python
print("\nDataset Information:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
filtered1 = df[df['sepal_length'] > 5.0]
print("\nFlowers with Sepal Length > 5.0:")
print(filtered1.head())
```

```python
setosa = df[df['species'] == 'setosa']
print("\nSetosa Flowers:")
print(setosa.head())
```

```python
filtered2 = df[df['petal_length'] > 4]
print("\nFlowers with Petal Length > 4:")
print(filtered2.head())
```

```python
group_species = df.groupby('species')

print("\nGrouped Data:")
print(group_species.size())
```

```python
print("\nMean values grouped by Species:")
print(group_species.mean())
```

```python
print("\nMaximum values grouped by Species:")
print(group_species.max())
```

```python
print("\nMinimum values grouped by Species:")
print(group_species.min())
```

```python
print("\nCount of each Species:")
print(df['species'].value_counts())
```

```python
agg_stats = df.groupby('species').agg(['mean','min','max','count'])
print("\nAggregate Statistics:")
print(agg_stats)
```

```python
# Convert sepal length column to numpy array
sepal_array = np.array(df['sepal_length'])

print("\nNumPy Mean of Sepal Length:", np.mean(sepal_array))
print("NumPy Standard Deviation:", np.std(sepal_array))
```
