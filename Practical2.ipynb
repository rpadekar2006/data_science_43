````python
# Simple A* Search Program

# Small graph with costs
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 1},
    'C': {'F': 2},
    'D': {},
    'F': {}
}

# Heuristic values (estimate to reach goal F)
heuristic = {
    'A': 4,
    'B': 2,
    'C': 1,
    'D': 1,
    'F': 0
}

def a_star(start, goal):

    open_list = [start]        # nodes to explore
    g_cost = {start: 0}        # cost from start
    parent = {}                # to store path

    while open_list:

        # Select node with smallest f = g + h
        current = open_list[0]
        for node in open_list:
            if g_cost[node] + heuristic[node] < g_cost[current] + heuristic[current]:
                current = node

        # If goal is reached
        if current == goal:
            path = []
            while current in parent:
                path.append(current)
                current = parent[current]
            path.append(start)
            path.reverse()
            return path

        open_list.remove(current)

        # Check neighbors
        for neighbor in graph[current]:

            new_cost = g_cost[current] + graph[current][neighbor]

            if neighbor not in g_cost or new_cost < g_cost[neighbor]:
                g_cost[neighbor] = new_cost
                parent[neighbor] = current
                open_list.append(neighbor)

    return None


# Run the algorithm
result = a_star('A', 'F')

print("Shortest Path:", result)
````

````python
# Simple Minimax Algorithm Example

# Minimax function
def minimax(depth, node_index, is_max):

    # Leaf nodes (end of game tree)
    scores = [3, 5, 2, 9]

    # If maximum depth reached, return score
    if depth == 2:
        return scores[node_index]

    # If it is Maximizer's turn
    if is_max:
        best = -1000

        # Check both children
        for i in range(2):
            value = minimax(depth + 1, node_index * 2 + i, False)
            best = max(best, value)

        return best

    # If it is Minimizer's turn
    else:
        best = 1000

        # Check both children
        for i in range(2):
            value = minimax(depth + 1, node_index * 2 + i, True)
            best = min(best, value)

        return best


# Start Minimax from root
result = minimax(0, 0, True)

print("Best value for the Maximizer is:", result)
````

````python
# Define the graph using an adjacency list (dictionary) - using the same graph as BFS example
graph_dfs = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

def dfs_recursive(graph, start_node, visited=None, traversal_order=None):
    # Initialize visited set and traversal_order list if they are not provided (first call)
    if visited is None:
        visited = set()
    if traversal_order is None:
        traversal_order = []

    # Mark the current node as visited
    visited.add(start_node)
    # Add the current node to the traversal order
    traversal_order.append(start_node)

    # Recursively visit all unvisited neighbors
    for neighbor in graph[start_node]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited, traversal_order)

    return traversal_order

# Define the starting node for the DFS
start_node_dfs = 'A'

# Perform DFS and get the traversal order
dfs_result = dfs_recursive(graph_dfs, start_node_dfs)

# Print the result
print(f"DFS traversal starting from node {start_node_dfs}: {dfs_result}")

# Show a sample output for a different start node
print("\n--- Another example ---")
start_node_dfs_2 = 'C'
dfs_result_2 = dfs_recursive(graph_dfs, start_node_dfs_2)
print(f"DFS traversal starting from node {start_node_dfs_2}: {dfs_result_2}")
````

### What is Depth-First Search (DFS)?

Depth-First Search (DFS) is another algorithm for traversing or searching tree or graph data structures. Unlike BFS, which explores level by level, DFS explores as far as possible along each branch before backtracking. Imagine exploring a maze by going down one path until you hit a dead end, then returning to the last junction and trying another path.

---

### How the Code Works:

1.  **Graph Representation**:
    *   We are using the same simple graph represented as a Python dictionary (`graph_dfs`) as in the BFS example. Each key is a node, and its value is a list of its direct neighbors.

2.  **`dfs_recursive(graph, start_node, visited=None, traversal_order=None)` Function**:
    *   **`visited = set()`**: This set keeps track of nodes we've already seen to prevent infinite loops and redundant processing.
    *   **`traversal_order = []`**: This list stores the nodes in the order they are visited by the DFS algorithm.
    *   The `visited` set and `traversal_order` list are passed recursively to maintain their state across function calls. They are initialized only on the very first call.

3.  **The Algorithm Steps (Recursive Logic)**:
    *   **Mark and Record**: When `dfs_recursive` is called for a `start_node`:
        *   The `start_node` is immediately added to the `visited` set.
        *   It's also added to the `traversal_order` list.
    *   **Explore Neighbors**: The function then looks at all `neighbor` nodes connected to the `start_node`.
    *   **Recursive Call**: For each `neighbor` that has *not* been `visited` yet:
        *   It makes a **recursive call** to `dfs_recursive(graph, neighbor, visited, traversal_order)`. This means it 'dives' deeper into the graph from that neighbor, exploring all paths from there before returning.
    *   **Backtracking**: Once all neighbors of the current `start_node` have been explored (either directly or through recursive calls), the function returns to the caller (the node that called it), effectively 'backtracking' up the current path.

4.  **Result**: The recursion continues until all reachable nodes from the initial `start_node` have been visited. The `traversal_order` list will then contain the nodes in the order they were discovered, reflecting the depth-first exploration.

### Sample Output:

```
DFS traversal starting from node A: ['A', 'B', 'D', 'E', 'C', 'F', 'G']
--- Another example ---
DFS traversal starting from node C: ['C', 'F', 'G']
```

````python
# Define the graph using an adjacency list (dictionary)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

def bfs(graph, start_node):
    # Keep track of visited nodes to avoid cycles and redundant processing
    visited = set()
    # Initialize a queue for BFS traversal
    # We'll use a list as a queue, adding to the end and removing from the beginning
    queue = []

    # Add the start node to the queue and mark it as visited
    queue.append(start_node)
    visited.add(start_node)

    # List to store the order of nodes visited
    traversal_order = []

    # Continue as long as there are nodes in the queue
    while queue:
        # Get the first node from the queue (dequeue operation)
        current_node = queue.pop(0) # Removing from the beginning of a list is simple for demonstration
        traversal_order.append(current_node)

        # Explore all neighbors of the current node
        # For each neighbor in the graph's adjacency list for current_node
        for neighbor in graph[current_node]:
            # If the neighbor has not been visited yet
            if neighbor not in visited:
                # Mark it as visited
                visited.add(neighbor)
                # Add it to the queue to be processed later
                queue.append(neighbor)

    return traversal_order

# Define the starting node for the BFS
start_node = 'A'

# Perform BFS and get the traversal order
bfs_result = bfs(graph, start_node)

# Print the result
print(f"BFS traversal starting from node {start_node}: {bfs_result}")

# Show a sample output for a different start node
print("\n--- Another example ---")
start_node_2 = 'C'
bfs_result_2 = bfs(graph, start_node_2)
print(f"BFS traversal starting from node {start_node_2}: {bfs_result_2}")
````

### What is Breadth-First Search (BFS)?

Breadth-First Search (BFS) is an algorithm for traversing or searching tree or graph data structures. It explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level. Think of it like exploring a maze one "layer" at a time, starting from your current position. You check all immediate paths, then all paths one step further, and so on.

---

### How the Code Works:

1.  **Graph Representation**:
    *   We represent our simple graph using a Python dictionary called `graph`. Each key in the dictionary is a node, and its value is a list of its direct neighbors. For example, `graph['A'] = ['B', 'C']` means node 'A' is connected to 'B' and 'C'.

2.  **`bfs(graph, start_node)` Function**:
    *   **`visited = set()`**: This is a `set` (a collection of unique items) used to keep track of all the nodes we've already seen. This is crucial to avoid getting stuck in loops in the graph and to make sure we process each node only once.
    *   **`queue = []`**: This is our queue, a list in this case. In BFS, we process nodes in the order they were added (First-In, First-Out, or FIFO). We add new nodes to the end of the list (`append()`) and take nodes from the beginning of the list (`pop(0)`).
    *   **`traversal_order = []`**: This list will store the nodes in the exact order they are visited by our BFS algorithm.

3.  **The Algorithm Steps**:
    *   **Start**: We begin by adding our `start_node` (e.g., 'A') to both the `queue` and the `visited` set.
    *   **Loop `while queue:`**: The core of BFS runs as long as there are nodes waiting in our queue to be explored.
    *   **`current_node = queue.pop(0)`**: We take the very first node from the queue. This is the node we are currently "exploring".
    *   **`traversal_order.append(current_node)`**: We add this `current_node` to our `traversal_order` list to record that we've processed it.
    *   **Explore Neighbors**: We then look at all the `neighbor` nodes connected to our `current_node`.
    *   **`if neighbor not in visited:`**: For each neighbor, we check if we've already seen it. If not:
        *   **`visited.add(neighbor)`**: We mark the neighbor as visited.
        *   **`queue.append(neighbor)`**: We add the neighbor to the end of our queue, meaning it will be processed after all the other nodes currently in the queue.

4.  **Result**: Once the `queue` is empty, it means we've explored all reachable nodes from the `start_node`, and the `traversal_order` list contains the nodes in the order they were discovered, level by level.

BFS traversal starting from node A: ['A', 'B', 'C', 'D', 'E', 'F', 'G']
