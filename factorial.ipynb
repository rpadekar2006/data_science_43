```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

```python
def multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Example usage:
number = 7
multiplication_table(number)

print("\n") # Add a newline for better readability between examples

number = 12
multiplication_table(number)
```

## Set up Information

### Subtask:
Define a small set of facts and rules for a very basic decision-making scenario, such as simple weather advice, using Python dictionaries.


**Reasoning**:
To set up the information for the decision-making scenario, I need to define the `knowledge_base` dictionary with current weather facts and a `rules` list of dictionaries, each containing `conditions` and `advice` as specified in the instructions.



```python
knowledge_base = {
    'temperature': 'mild',
    'precipitation': 'none',
    'wind': 'low',
    'sky': 'sunny'
}

rules = [
    {
        'conditions': {'temperature': 'hot', 'precipitation': 'none'},
        'advice': 'Wear light clothes, a hat, and sunglasses.'
    },
    {
        'conditions': {'temperature': 'cold', 'wind': 'high'},
        'advice': 'Wear a warm coat, gloves, and a scarf. Consider wind-resistant outer layers.'
    },
    {
        'conditions': {'precipitation': 'rainy'},
        'advice': 'Take an umbrella and wear a waterproof jacket.'
    },
    {
        'conditions': {'temperature': 'mild', 'sky': 'cloudy'},
        'advice': 'A light jacket or sweater might be comfortable.'
    },
    {
        'conditions': {'temperature': 'cold', 'precipitation': 'snowy'},
        'advice': 'Wear a heavy winter coat, snow boots, and thermal layers. Stay warm!'
    }
]

print("Knowledge Base:", knowledge_base)
print("\nRules:")
for i, rule in enumerate(rules):
    print(f"Rule {i+1}: {rule}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage:
num1 = 7
print(check_odd_even(num1))

num2 = 10
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

## Build Decision Maker (Basic)

### Subtask:
Create a simple function to check the facts against the rules and return a basic decision.


**Reasoning**:
The subtask requires creating a function to evaluate facts against rules and provide advice. This code block will define the `make_decision` function as described in the instructions.



```python
def make_decision(facts, rules):
    for rule in rules:
        conditions_met = True
        for fact_key, fact_value in rule['conditions'].items():
            if fact_key not in facts or facts[fact_key] != fact_value:
                conditions_met = False
                break
        if conditions_met:
            return rule['advice']
    return 'No specific advice for these conditions.'

# Test the function with the existing knowledge_base and rules
advice = make_decision(knowledge_base, rules)
print("Current Facts:", knowledge_base)
print("Generated Advice:", advice)
```
