```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Fixed values (change these as needed)
mu = 0          # Mean
sigma = 1       # Standard Deviation
n = 1000        # Number of samples

# Generate random data
data = np.random.normal(mu, sigma, n)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6)

# Plot PDF curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, linewidth=2)

plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```

```python
import numpy as np

# User Input
n = int(input("Enter number of elements: "))
data = []

for i in range(n):
    value = float(input(f"Enter element {i+1}: "))
    data.append(value)

data = np.array(data)

# Calculate statistics
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("\nData:", data)
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import numpy as np

# User Input
tosses = int(input("Enter number of coin tosses: "))

# Simulate coin toss
results = np.random.choice(['H', 'T'], size=tosses)

# Count outcomes
heads = np.count_nonzero(results == 'H')
tails = np.count_nonzero(results == 'T')

# Display results
print("\nTotal Tosses:", tosses)
print("Heads:", heads)
print("Tails:", tails)

print("Probability of Head:", heads / tosses)
print("Probability of Tail:", tails / tosses)
```
