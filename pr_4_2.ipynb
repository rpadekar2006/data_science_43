```python
import pandas as pd
import numpy as np

# Example dataset (you can replace this with Titanic dataset column like Age or Fare)
data = [22, 38, 26, 35, 28, 42, 19, 30]

# Convert to pandas Series
dataset = pd.Series(data)

# Mean
mean_value = dataset.mean()

# Median
median_value = dataset.median()

# Variance
variance_value = dataset.var()

# Standard Deviation
std_dev_value = dataset.std()

# Print results
print("Mean:", mean_value)
print("Median:", median_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_dev_value)
```
